## Filter A DataFrame  Based On A Condition
- Pandas needs a **Series** of Booleans to perform a filter.
- Pass the Boolean Series inside square brackets after the **DataFrame**.
- We can generate a Boolean Series using a wide variety of operations (equality, inequality, less than, greater than, inclusion, etc)

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
employees["Gender"] == "Male"
# Para filtrar los registros que cumplen con la condición se debe pasar la serie de booleanos al DataFrame
employees[employees["Gender"] == "Male"]
```

```python
employees["Team"] == "Finance"
employees[employees["Team"] == "Finance"]
```

```python
# Una mejor práctica es guardar la serie de booleanos en una variable
on_finance_team = employees["Team"] == "Finance"
employees[on_finance_team]
```

```python
import datetime as dt
```

```python
is_senior_management = employees["Senior Management"]
employees[is_senior_management]

up_to_110K = employees["Salary"] > 110000
employees[up_to_110K]

bonus_1_5 = employees["Bonus %"] < 1.5
employees[bonus_1_5]

old_employees = employees["Start Date"] < "1985-01-01"
employees[old_employees]

employees[employees["Last Login Time"] < dt.time(12, 0, 0)]
```

## Filter with More than One Condition (AND)
- Add the `&` operator in between two Boolean **Series** to filter by multiple conditions.
- We can assign the **Series** to variables to make the syntax more readable.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
# female employees who work in Marketing who earn over $100k a year
is_female = employees["Gender"] == "Female"

is_in_marketing_team = employees["Team"] == "Marketing"

salary_over_100k_year = employees["Salary"] > 100000

is_female & is_in_marketing_team & salary_over_100k_year
```

```python
employees[is_female & is_in_marketing_team & salary_over_100k_year]
```

## Filter with More than One Condition (OR)
- Use the `|` operator in between two Boolean **Series** to filter by *either* condition.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
# Employees who are either senior management OR started before January 1st, 1990
is_senior_management = employees["Senior Management"]

started_before_1990 = employees["Start Date"] < "1990-01-01"

is_senior_management | started_before_1990
```

```python
employees[is_senior_management | started_before_1990]
```

```python
# First Name is Robert who work in Client Services OR Start Date after 2016-06-01
is_robert = employees["First Name"] == "Robert"
is_client_services_team = employees["Team"] == "Client Services"
after_2016_06_01 = employees["Start Date"] > "2016-06-01"

condition = ((is_robert & is_client_services_team) | after_2016_06_01)
```

```python
employees[condition]
```

## The isin Method
- The `isin` **Series** method accepts a collection object like a list, tuple, or **Series**.
- The method returns True for a row if its value is found in the collection.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees.head()
```

```python
# Legal Team or Sales Team or Prodcut Team
is_legal_team = employees["Team"] == "Legal"
is_sales_team = employees["Team"] == "Sales"
is_product_team = employees["Team"] == "Product"

condition = is_legal_team | is_sales_team | is_product_team

employees[condition]
```

## The isnull and notnull Methods

## The duplicated Method
- The `duplicated` method returns True if a **Series** value is a duplicate.
- Pandas will mark one occurrence of a repeated value as a non-duplicate.
- Use the `keep` parameter to designate whether the first or last occurrence of a repeated value should be considered the "non-duplicate".
- Pass False to the `keep` parameter to mark all occurrences of repeated values as duplicates.
- Use the tilde symbol (`~`) to invert a **Series's** values. Trues will become Falses, and Falses will become trues.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees
```

El comportamiento predeterminado de `duplicated` es marcar la primera ocurrencia de un valor duplicado como no duplicado. Por lo tanto, si un valor se repite, la primera ocurrencia se considera no duplicada y las siguientes se consideran duplicadas.

```python
employees["First Name"].duplicated()
```

keep='first':

- La primera aparición NO se marca como duplicada (False).
- Todas las siguientes repeticiones se marcan como duplicadas (True).

```python
# Esto genera un DataFrame con los registros duplicados en la columna "First Name" pero solo muestra la primera ocurrencia
employees[employees["First Name"].duplicated()]
```

keep='last':

- La última aparición NO se marca como duplicada (False).
- Todas las anteriores se marcan como duplicadas (True).


```python
# Esto genera un DataFrame con los registros duplicados en la columna "First Name" pero solo muestra el primer registro duplicado
employees[employees["First Name"].duplicated(keep="last")]
```

```python
# Si se quiere mostrar todos los registros duplicados se puede usar el parámetro keep=False
employees[employees["First Name"].duplicated(keep=False)]
```

```python
# Si se quiere mostrar todos los registros no duplicados se puede usar el parámetro keep=False y negar la condición
employees[~employees["First Name"].duplicated(keep=False)]
```

## The drop_duplicates Method
- The `drop_duplicates` method deletes rows with duplicate values.
- By default, it will remove a row if *all* of its values are shared with another row.
- The `subset` parameter configures the columns to look for duplicate values within.
- Pass a list to `subset` parameter to look for duplicates across multiple columns.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees
```

El método `drop_duplicates` elimina las filas con valores duplicados. Por defecto, eliminará una fila si *todos* sus valores se comparten con otra fila.

```python
employees.drop_duplicates()
employees.drop_duplicates("Team")
```

El parametro keep='first' es el valor predeterminado, por lo que la primera ocurrencia de un valor duplicado se conserva y las siguientes se eliminan.

El parametro keep='last' conserva la última ocurrencia de un valor duplicado y elimina las anteriores.

```python
employees.drop_duplicates("Team", keep="first")
employees.drop_duplicates("Team", keep="last")
```

El parametro keep=False de `drop_duplicates` elimina todas las filas duplicadas, independientemente de si son la primera o la última aparición.

```python
employees.drop_duplicates("Team", keep=False)
```

```python
employees.drop_duplicates("First Name", keep=False)
```

Al enviar una lista de columnas al parametro `subset`, `drop_duplicates` buscará valores duplicados solo en las columnas especificadas.

```python
employees.drop_duplicates(["Senior Management", "Team"]).sort_values("Team")
employees.drop_duplicates(["Senior Management", "Team"], keep="last").sort_values("Team")
```

## The unique and nunique Methods
- The `unique` method on a **Series** returns a collection of its unique values. The method does not exist on a **DataFrame**.
- The `nunique` method returns a *count* of the number of unique values in the **Series**/**DataFrame**.
- The `dropna` parameter configures whether to include or exclude missing (`NaN`) values.

```python
employees = pd.read_csv("employees.csv", parse_dates=["Start Date"], date_format="%m/%d/%Y")
employees["Last Login Time"] = pd.to_datetime(employees["Last Login Time"], format="%H:%M %p").dt.time
employees["Senior Management"] = employees["Senior Management"].astype(bool)
employees["Gender"] = employees["Gender"].astype("category")
employees
```

```python
employees["Gender"].unique()
```

El tipo de retorno de `unique` es un objeto de tipo `ndarray` de NumPy, que es una matriz unidimensional de valores únicos. Varía en longitud según la cantidad de valores únicos en la Serie.

El método nunique() devuelve el número de valores únicos en una serie o columna de un DataFrame y excluye los valores NaN.

```python
employees["Gender"].nunique()
```

```python
# El parámetro dropna=False permite contar los valores nulos
employees["Gender"].nunique(dropna=False)
```

El método unique() no tiene un equivalente en un DataFrame. Solo se puede llamar en una Serie.

```python
# El método unique() también se puede usar en un DataFrame y devuelve una serie con los valores únicos de cada columna
employees.nunique()
employees.nunique(dropna=False)
```
